\begin{equation*}
\beta_h(V) = \frac{1}{1 + e^{(-.1(V+35))}}
\end{equation*}

\begin{equation*}
\tau_p(V) = \frac{1}{\alpha_p(V) + \beta_p(V)}
\end{equation*}

\begin{equation*}
p_{\infty} = \frac{\alpha_p(V)}{\alpha_p(V) + \beta_p(V)}
\end{equation*}

\begin{equation*}
V_{\infty} = \frac{g_LE_L + g_Kn^4E_K + g_{Na}m^3hE_{na} + \frac{I}{A}}{g_L + g_Kn^4 + g_{Na}m^3h}
\end{equation*}

\begin{equation*}
p(t + \Delta t) = p_{\infty} + (p - p_{\infty})e^{\frac{-\Delta t}{\tau_p}}
\end{equation*}
